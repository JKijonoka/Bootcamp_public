```python
# For this example we will create a dataframe with 4 rows and 4 columns (4*4 matrix)
data = pd.read_csv('nba.csv')
data.head()
```

```python
import joblib
import pandas as pd
from sklearn.preprocessing import StandardScaler

NUM_COLS = ["Number", "Age", "Weight", "Salary"]  # define explicitly (recommended)

scaler = StandardScaler()
scaler.fit(data[NUM_COLS])

artifact = {
    "scaler": scaler,
    "feature_names": NUM_COLS,     # order matters!
    "version": 1
}

joblib.dump(artifact, "scaler_artifact.joblib")
```

```python
# in different space of work
import joblib
import pandas as pd

def apply_scaler(df: pd.DataFrame, artifact_path: str, inverse: bool = False) -> pd.DataFrame:
    art = joblib.load(artifact_path)
    scaler = art["scaler"]
    cols = art["feature_names"]

    # Ensure required columns exist
    missing = [c for c in cols if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns for scaling: {missing}")

    # Align column order exactly
    X = df[cols]

    # Transform
    X_out = scaler.inverse_transform(X) if inverse else scaler.transform(X)
    print(X_out)

    # Put back into a copy of df
    out = df.copy()
    out[cols] = X_out
    return out
```
